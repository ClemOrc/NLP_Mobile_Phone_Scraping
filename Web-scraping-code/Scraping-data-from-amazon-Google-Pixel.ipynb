```python
import csv
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
import time

# Setup Selenium WebDriver
chrome_options = Options()
chrome_options.add_argument("--headless")  # Run in headless mode
chrome_options.add_argument("--disable-gpu")
chrome_options.add_argument("--no-sandbox")
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=chrome_options)

# Open Amazon product page with reviews
url = "https://www.amazon.ca/Google-Pixel-Pro-Unlocked-Smartphone/dp/B0D8V7CHDZ"
driver.get(url)
time.sleep(3)  # Allow the page to load

# Extract product title and brand
try:
    product_title = driver.find_element(By.ID, "productTitle").text.strip()
    product_brand = driver.find_element(By.ID, "bylineInfo").text.strip()
    print(f"Product Title: {product_title}")
    print(f"Product Brand: {product_brand}")
except Exception as e:
    print(f"Error extracting product details: {e}")
    product_title = "N/A"
    product_brand = "N/A"

# Function to extract reviews from the same page
def extract_reviews():
    reviews = []
    review_blocks = driver.find_elements(By.XPATH, '//div[starts-with(@id, "customer_review")]')

    for block in review_blocks:
        try:
            # Extract reviewer name
            reviewer_name = block.find_element(By.CLASS_NAME, "a-profile-name").text.strip()

            # Extract review date
            review_date = block.find_element(By.XPATH, './/span[@data-hook="review-date"]').text.strip()

            # Extract review title
            review_title = block.find_element(By.XPATH, './/a[@data-hook="review-title"]').text.strip()

            # Extract review body (expand if necessary)
            try:
                read_more = block.find_element(By.CLASS_NAME, "a-expander-header")
                driver.execute_script("arguments[0].click();", read_more)  # Expand "Read More"
                time.sleep(0.5)  # Wait for expansion
            except:
                pass  # No "Read More" link

            review_body = block.find_element(By.XPATH, './/span[@data-hook="review-body"]').text.strip()

            # Append the review details
            reviews.append({
                "Product Title": product_title,
                "Product Brand": product_brand,
                "Reviewer Name": reviewer_name,
                "Review Date": review_date,
                "Review Title": review_title,
                "Review Body": review_body,
            })
        except Exception as e:
            print(f"Error extracting a review: {e}")

    return reviews

# Extract reviews
reviews = extract_reviews()

# Save reviews to CSV
output_file = "amazon_reviews_same_page.csv"
try:
    with open(output_file, mode="w", newline="", encoding="utf-8") as file:
        fieldnames = ["Product Title", "Product Brand", "Reviewer Name", "Review Date", "Review Title", "Review Body"]
        writer = csv.DictWriter(file, fieldnames=fieldnames)
        writer.writeheader()
        writer.writerows(reviews)
    print(f"Reviews successfully saved to {output_file}.")
except Exception as e:
    print(f"Error saving reviews to CSV: {e}")

# Close the browser
driver.quit()
```
